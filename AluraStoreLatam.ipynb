### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
#Pasando a DataFrame los precios de todas las tiendas.
df = pd.DataFrame(tienda)
df2 = pd.DataFrame(tienda2)
df3 = pd.DataFrame(tienda3)
df4 = pd.DataFrame(tienda4)
```

```python
totales = [
    df['Precio'].sum(),
    df2['Precio'].sum(),
    df3['Precio'].sum(),
    df4['Precio'].sum(),
]

def formatocosto(val):
  return f'${val:,.2f}'

ingresototal = 0
for i in range(len(totales)):
  ingresototal += totales[i]
  print(f'Ingreso total de la tienda {i+1} es: {formatocosto(totales[i])}')

print(f"El ingreso total sumando todas las tiendas es: {formatocosto(ingresototal)}")
```

```python
import matplotlib.pyplot as plt

dic_totales = {
    'Tienda 1': df['Precio'].sum(),
    'Tienda 2': df2['Precio'].sum(),
    'Tienda 3': df3['Precio'].sum(),
    'Tienda 4': df4['Precio'].sum(),

}


plt.figure(figsize=(10, 6))
plt.bar(dic_totales.keys(), dic_totales.values(), color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])
plt.title('Ingresos Totales por Tienda', fontweight='bold')
plt.xlabel('Tiendas')
plt.ylabel('Ingresos')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.gca().yaxis.set_major_formatter('${x:,.0f}')

plt.show()
```

# 2. Ventas por categoría

Debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
tiendas_cat = [df,df2,df3,df4]

ventas_por_tienda = pd.concat(
    [d['Categoría del Producto'].value_counts() for d in tiendas_cat],
    axis  = 1,
    keys=['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
)

print(ventas_por_tienda)
max_por_tienda = ventas_por_tienda.idxmax()
resultados = ventas_por_tienda.max()


for i, (tienda,categoria) in enumerate(max_por_tienda.items(),1):
  print(f'La categoría más vendida en la {tienda} es: {categoria} con {resultados[tienda]:.0f} ventas')
```

```python
ventas_por_tienda.plot(kind='bar', figsize=(12, 6), width=0.8)
plt.title('Ventas por Categoría en Cada Tienda', fontweight='bold')
plt.xlabel('Categoría')
plt.ylabel('Cantidad de Ventas')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Tiendas')
plt.show()
```

# 3. Calificación promedio de la tienda


```python
promedios = [
    df['Calificación'].mean(),
    df2['Calificación'].mean(),
    df3['Calificación'].mean(),
    df4['Calificación'].mean(),
]

for i in range(len(promedios)):
  print(f'La calificación promedio de la tienda {i+1} es: {promedios[i]:.2f}')

promedio_general = sum(promedios)/ len(promedios)

print(f'El promedio general entre las 4 tiendas es: {promedio_general:.2f}')
```

```python
dic_promedios = {
    'Tienda 1':df['Calificación'].mean(),
    'Tienda 2':df2['Calificación'].mean(),
    'Tienda 3':df3['Calificación'].mean(),
    'Tienda 4':df4['Calificación'].mean(),
}

plt.figure(figsize=(10, 4))
plt.plot(dic_promedios.keys(), dic_promedios.values(), marker='o', linestyle='-', color='#2ca02c', linewidth=2)
plt.title('Tendencia de Calificación Promedio por Tienda', fontweight='bold')
plt.xlabel('Tiendas')
plt.ylabel('Calificación (Promedio)')
plt.ylim(3.9, 4.1)
plt.grid(axis='y', linestyle='--')


for i, val in enumerate(promedios):
    plt.text(i, val + 0.01, f'{val:.2f}', ha='center', fontweight='bold')

plt.show()
```

# 4. Productos más y menos vendidos

En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python

df['Tienda'] = 'Tienda 1'
df2['Tienda'] = 'Tienda 2'
df3['Tienda'] = 'Tienda 3'
df4['Tienda'] = 'Tienda 4'


df_total = pd.concat([df,df2,df3,df4])

df_total['Precio'] = df_total['Precio'].astype(float)
df_total['Cantidad de cuotas'] = df_total['Cantidad de cuotas'].astype(int)

df_total['Ingreso'] = df_total['Precio']

ventas = df_total.groupby(['Tienda', 'Producto'])['Ingreso'].sum().reset_index()

def top_productos(tienda):
  data = ventas[ventas['Tienda']==tienda]

  mas_vendidos = data.sort_values(by='Ingreso',ascending=False).head(3)
  menos_vendidos = data.sort_values(by='Ingreso',ascending=True).head(3)

  print(f'\n Los 3 productos más vendidos de {tienda}:')
  for _, fila in mas_vendidos.iterrows():
    print(f"  {fila['Producto']} - Ingresos: {formatocosto(fila['Ingreso'])}")

  print(f'\n Los 3 productos menos vendidos de {tienda}:')
  for _, fila in menos_vendidos.iterrows():
    print(f"  {fila['Producto']} - Ingresos: {formatocosto(fila['Ingreso'])}")

tiendas_unicas = ventas['Tienda'].unique()

for tienda in tiendas_unicas:
  top_productos(tienda)
```

```python
def top_productos(tienda):
    data = ventas[ventas['Tienda'] == tienda]
    mas_vendidos = data.sort_values(by='Ingreso', ascending=False).head(3)
    return mas_vendidos

tiendas_unicas = ventas['Tienda'].unique()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Top 3 Productos Más Vendidos por Tienda', fontweight='bold', y=1.02)

for ax, tienda in zip(axes.flatten(), tiendas_unicas):
    mas_vendidos = top_productos(tienda)
    ax.barh(
        mas_vendidos['Producto'],
        mas_vendidos['Ingreso'],
        color=['#2ca02c', '#1f77b4', '#ff7f0e']
    )
    ax.set_title(tienda)
    ax.set_xlabel('Ingresos')
    ax.grid(axis='x', linestyle='--', alpha=0.6)

    for ing, prod in zip(mas_vendidos['Ingreso'], mas_vendidos['Producto']):
        ax.text(ing, prod, f'${ing/1e6:.1f}M', va='center', ha='left', fontsize=9)

plt.tight_layout()
plt.show()
```

```python
def top_productos(tienda):
    data = ventas[ventas['Tienda'] == tienda]
    menos_vendidos = data.sort_values(by='Ingreso', ascending=True).head(3)
    return menos_vendidos

tiendas_unicas = ventas['Tienda'].unique()

fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Top 3 Productos Menos Vendidos por Tienda', fontweight='bold', y=1.02)

for ax, tienda in zip(axes.flatten(), tiendas_unicas):
    menos_vendidos = top_productos(tienda)
    ax.barh(
        menos_vendidos['Producto'],
        menos_vendidos['Ingreso'],
        color=['#ff9896', '#ff7f0e', '#1f77b4']
    )
    ax.set_title(tienda)
    ax.set_xlabel('Ingresos')
    ax.grid(axis='x', linestyle='--', alpha=0.6)

    ax.set_xlim(0, menos_vendidos['Ingreso'].max() * 1.1)

    for ing, prod in zip(menos_vendidos['Ingreso'], menos_vendidos['Producto']):
        ax.text(ing, prod, f'${ing:.2f}', va='center', ha='left', fontsize=9)

plt.tight_layout()
plt.show()
```
