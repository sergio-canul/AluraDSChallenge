### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
#Pasando a DataFrame los precios de todas las tiendas.
df = pd.DataFrame(tienda)
df2 = pd.DataFrame(tienda2)
df3 = pd.DataFrame(tienda3)
df4 = pd.DataFrame(tienda4)
```

```python
"""Tabla Tienda 1"""
df.head()
```

```python
"""Tabla Tienda 2"""
df2.head()
```

```python
"""Tabla Tienda 3"""
df3.head()
```

```python
"""Tabla Tienda 4"""
df4.head()
```

```python
totales = [
    df['Precio'].sum(),
    df2['Precio'].sum(),
    df3['Precio'].sum(),
    df4['Precio'].sum(),
]

def formatocosto(val):
  return f'${val:,.2f}'

ingresototal = 0
for i in range(len(totales)):
  ingresototal += totales[i]
  print(f'Ingreso total de la tienda {i+1} es: {formatocosto(totales[i])}')

print(f"El ingreso total sumando todas las tiendas es: {formatocosto(ingresototal)}")
```

# 2. Ventas por categoría

Debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
tiendas_cat = [df,df2,df3,df4]

ventas_por_tienda = pd.concat(
    [d['Categoría del Producto'].value_counts() for d in tiendas_cat],
    axis  = 1,
    keys=['Tienda 1','Tienda 2','Tienda 3','Tienda 4']
)

print(ventas_por_tienda)
max_por_tienda = ventas_por_tienda.idxmax()
resultados = ventas_por_tienda.max()


for i, (tienda,categoria) in enumerate(max_por_tienda.items(),1):
  print(f'La categoría más vendida en la {tienda} es: {categoria} con {resultados[tienda]:.0f} ventas')

```
